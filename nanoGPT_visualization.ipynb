```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Vijay data
df_vijay = pd.read_csv('results_vijay.csv')

# Sort by experiment number extracted from the experiment string index
df_vijay['Exp Num'] = df_vijay.index + 1

# 1. Train and Val Loss Curves
plt.figure(figsize=(18, 6))
plt.plot(df_vijay['Exp Num'], df_vijay['Train Loss'], marker='o', label='Train Loss')
plt.plot(df_vijay['Exp Num'], df_vijay['Val Loss'], marker='x', label='Validation Loss')
plt.xticks(df_vijay['Exp Num'])
plt.title('Loss Comparison for Vijay Experiments (1–32)')
plt.xlabel('Experiment ID')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# 2. Train–Val Gap Curve
plt.figure(figsize=(18, 6))
plt.plot(df_vijay['Exp Num'], df_vijay['Loss Gap'], marker='d', color='orange', label='Loss Gap')
plt.axhline(0.10, linestyle='--', color='red', label='High Overfitting Threshold')
plt.xticks(df_vijay['Exp Num'])
plt.title('Train–Validation Gap for Vijay Experiments')
plt.xlabel('Experiment ID')
plt.ylabel('Gap')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# 3. Best experiments by validation loss
print("\nTop 10 Vijay Experiments by Lowest Validation Loss:")
display(df_vijay.sort_values(by='Val Loss').head(10))

# 4. Best experiments by lowest gap (generalization)
print("\nTop 10 Vijay Experiments by Lowest Gap:")
display(df_vijay.sort_values(by='Loss Gap').head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Raniya data
df_raniya = pd.read_csv('results_raniya.csv')

# Assign experiment numbers (65–96)
df_raniya['Exp Num'] = range(65, 65 + len(df_raniya))

# 1. Train and Val Loss Curves
plt.figure(figsize=(18, 6))
plt.plot(df_raniya['Exp Num'], df_raniya['Train Loss'], marker='o', label='Train Loss')
plt.plot(df_raniya['Exp Num'], df_raniya['Val Loss'], marker='x', label='Validation Loss')
plt.xticks(df_raniya['Exp Num'], rotation=90)
plt.title('Loss Comparison for Raniya Experiments (65–96)')
plt.xlabel('Experiment ID')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# 2. Train–Val Gap Curve
plt.figure(figsize=(18, 6))
plt.plot(df_raniya['Exp Num'], df_raniya['Loss Gap'], marker='d', color='orange', label='Loss Gap')
plt.axhline(0.10, linestyle='--', color='red', label='High Overfitting Threshold')
plt.xticks(df_raniya['Exp Num'], rotation=90)
plt.title('Train–Validation Gap for Raniya Experiments')
plt.xlabel('Experiment ID')
plt.ylabel('Gap')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# 3. Best experiments by validation loss
print("\nTop 10 Raniya Experiments by Lowest Validation Loss:")
display(df_raniya.sort_values(by='Val Loss').head(10))

# 4. Best experiments by lowest gap
print("\nTop 10 Raniya Experiments by Lowest Gap:")
display(df_raniya.sort_values(by='Loss Gap').head(10))
```
